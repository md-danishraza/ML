#### DBSCAN

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, DBSCAN
```

sample dataset

```python
X, y = make_moons(n_samples=300, noise=0.05, random_state=42)
```

```python
plt.figure(figsize=(12, 5))
```

K-Means tries to find centroids. It fails on crescents.

```python
kmeans = KMeans(n_clusters=2, random_state=42)
y_kmeans = kmeans.fit_predict(X)

plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
plt.title("K-Means (Fails on Moons)")
plt.xlabel("K-Means cuts linearly through the shape")
```

- eps=0.2: The radius of the neighborhood
- min_samples=5: Need 5 points to form a dense region

```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
y_dbscan = dbscan.fit_predict(X)
```

```python
plt.subplot(1, 2, 2)
# Note: Points colored -1 are Noise/Outliers (if any)
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='viridis', s=50)
plt.title("DBSCAN (Follows the Shape)")
plt.xlabel("DBSCAN flows along the density")

plt.show()
```
