#### hierarchical clustering

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
```

sample data

```python
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
plt.figure(figsize=(12, 5))

# 2. Draw the Dendrogram
# We use 'ward' linkage to minimize variance
plt.subplot(1, 2, 1)
plt.title("Dendrogram")
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.axhline(y=10, color='r', linestyle='--') # Drawing a cut line
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
```

Training model based on dendrogram <br>
longest vertical distance suggests 4 clusters.

```python
hc = AgglomerativeClustering(n_clusters=4, linkage='ward')
y_hc = hc.fit_predict(X)
```

```python
plt.subplot(1, 2, 2)
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], s=50, c='red', label='Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], s=50, c='blue', label='Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], s=50, c='green', label='Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], s=50, c='cyan', label='Cluster 4')

plt.title('Hierarchical Clustering (Agglomerative)')
plt.legend()
plt.tight_layout()
plt.show()
```
