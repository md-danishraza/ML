# Dimensionality Reduction - PCA

#### importing Libs

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# make plot inline
%matplotlib inline
```

#### Loading dataset - breastCancer

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset = load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset["DESCR"])
```

```python
cancer_dataset.target_names
```

#### creating a Dataframe

```python
df = pd.DataFrame(data=cancer_dataset['data'],columns=cancer_dataset["feature_names"])
df.head()
```

#### standardizing the features

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#  Calculates the parameters needed for transformation (e.g., mean and std for standard scaling) and stores them internally.
scaler.fit(df)
# Applies the transformation using those stored parameters.
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

#### observation
This is a direct consequence of the empirical rule (also called the 68–95–99.7 rule) for normal distributions:

| Range | % of Data (Approx.) |

| ( \mu \pm 1\sigma ) → (−1 to +1) | 68% | 

| ( \mu \pm 2\sigma ) → (−2 to +2) | 95% | 

| ( \mu \pm 3\sigma ) → (−3 to +3) | 99.7% | 

#### apply PCA 
30 feautures to only 2 features

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
data_pca = pca.fit_transform(scaled_data)
data_pca
```

```python
#### captured variance
pca.explained_variance_
```

#### Visualizing the new features (PC1 and PC2)

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_dataset['target'],cmap="plasma")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend()
plt.plot()
```
