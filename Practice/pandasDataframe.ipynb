### Read csv add new column save it back

```python
import pandas as pd
import numpy as np
```

```python


data = pd.read_csv("./students.csv")

data['newColumn'] = data['Grade'] * 100

data.to_csv("updated.csv")
```

#### drop,rename a column

```python
data = data.drop(columns=['newColumn'])
data
```

#### sorting data

```python
data.sort_values(by=['Grade'],ascending=False)
```

#### find min, max, mean and sum 

```python
min = data['Age'].min()
print(f"min age is = {min}")
max = data['Age'].max()
print(f"max age is = {max}")
mean = data['Age'].mean()
print(f"mean age is = {mean}")
data['Marks'] = [88, 92, 75, 85, 90,50,56,70,57,95]
totalMarks = data['Marks'].sum()
print(f"total marks  is = {totalMarks}")
```

#### handling missing values

```python
# first dropping Marks and adding with nan values
data.drop(columns=["Marks"])
data['Marks'] = [88, np.nan, 75, 85, np.nan,50,56,70,np.nan,95]

# handling it 

data['Marks_Dropped'] = data['Marks'].dropna()
data['Marks_Fill_Zero'] = data['Marks'].fillna(0)
data['Marks_Fill_Mean'] = data['Marks'].fillna(data['Marks'].mean().__round__(2))
data['Marks_Fill_Median'] = data['Marks'].fillna(data['Marks'].median())
data['Marks_Fill_Mode'] = data['Marks'].fillna(data['Marks'].mode()[0])
data
```

```python
# data = data.drop(columns=['Marks_Dropped','Marks_Fill_Zero','Marks_Fill_Mean','Marks_Fill_Median','Marks_Fill_Mode'])
data = pd.read_csv("./students.csv")
data['Marks'] = [88, np.nan, 75, 85, np.nan,50,56,70,np.nan,95]
data["Marks"] = data['Marks'].fillna(data['Marks'].mean().__round__(2))
```

#### Handling outliers

```python
data['outlierCol'] = [88, 92, 75, 85, 90, 50, 56, 70,80, 999]

Q1 = data['outlierCol'].quantile(0.25)
Q3 = data['outlierCol'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers_iqr = data[(data['outlierCol'] < lower_bound) | (data['outlierCol'] > upper_bound)]
print(outliers_iqr)
# getting indexes
indexes = outliers_iqr.index.to_list()
data = data.drop(index=indexes)
# resetting index
data.reset_index(drop=True, inplace=True)
data
```
