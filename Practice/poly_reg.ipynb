#### polynomial regression

 It is useful when the data exhibits a non-linear relationship allowing the model to fit a curve to the data.

##### importing libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

#### job vs salaries dataset

```python
url = "https://raw.githubusercontent.com/krishnaik06/Polynomial-" \
"inear-Regression/master/Position_Salaries.csv"
dataset = pd.read_csv(url)

print(dataset.head())
```

#### features and target - 2d

```python
X = dataset.iloc[:, 1:2].values   
y = dataset.iloc[:, 2].values     
```

#### fitting linear regression

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

#### fitting polynomial regression with degree of 4

```python
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)

lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

#### predicting and plotting simple linear regression

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X), color='blue')
plt.title('Linear Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

#### predicting and plotting polynomial regression

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color='red')
plt.plot(X_grid, lin_reg2.predict(poly_reg.fit_transform(X_grid)), color='blue')
plt.title('Polynomial Regression (degree=4)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# Predict using Linear Regression
print("Linear Prediction (Level 6.5):", lin_reg.predict([[6.5]]))

# Predict using Polynomial Regression
print("Polynomial Prediction (Level 6.5):", lin_reg2.predict(poly_reg.transform([[6.5]])))
```

```python
from sklearn.metrics import r2_score

y_pred_poly = lin_reg2.predict(X_poly)
print("R² Score (Polynomial):", r2_score(y, y_pred_poly))
```
